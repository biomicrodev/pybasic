# Scalar Shrinkage

```python
import numpy as np
import numpy.typing as npt


def scalar_shrink(arr: npt.NDArray, epsilon=0.1) -> npt.NDArray:
    return np.sign(arr) * np.maximum(np.abs(arr) - epsilon, 0)
```

```python
import perfplot

rng = np.random.default_rng(seed=0)

perfplot.show(
    setup=lambda n: rng.random(size=n),
    kernels=[scalar_shrink],
    n_range=[10**k for k in range(8)],
    xlabel='n elements'
)
```
